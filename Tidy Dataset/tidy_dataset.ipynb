Astronomical tides data are divided in two datasets by year. For a better management of the datasets, we unify both datasets in a single one in which observations are chronologically ordered.

```python
#Concat astro data
astro=pd.concat([astro1,astro2])
#Rename columns
astro=astro.rename(columns={"data":"Data"})
```

By the documentation provided by "Comune di Venezia", we understand that astronomical tides are measured in cm.\
Data about the observed sea level in all the other datasets are reported in metres.\
Therefore, we convert these observation in metres without any loss of significative figures.

```python
#Converting astro data from cm to m
astro['valore']=astro['valore']/100
astro
```

We merge all the datasets into a single one.

```python
#Merging data
from functools import reduce

dfs = [cavalli, burano, ismar1, ismar2, san_giorgio, miser, salute, faro, meteo, astro]
venice = reduce(lambda  left,right: pd.merge(left,right,on=['Data'],how='outer'), dfs).sort_values('Data').reset_index(drop=True)
venice
```

```python
#Selecting relevant period of time
venice=venice.loc[(venice['Data']>='2022-11-15 00:00:00') & (venice['Data']<='2023-01-31 23:55:00')].reset_index(drop=True)

venice
```

We hourly aggregate observations following these rules:
1. Observations already reported as an average, are aggregated using the mean.
2. Observations reported using the maximum, are aggregated using the max.
3. Observations about precipitations are aggregated using a sum.
4. All other observations are aggregated using the median.
5. In all of the above aggregation methods, missing values are not taken into consideration.

```python
venice=venice.resample('H', on='date').agg({'cavalli_atmospheric_pressure':'median', 'cavalli_air_temperature':'median',
                                     'cavalli_relative_humidity':'median','cavalli_solar_radiation':'median',
                                     'cavalli_precipitation':'sum','burano_sea_level':'mean','ismar1_average_sea_level':'mean', 
                                     'ismar1_average_wind_direction':'mean','ismar1_average_wind_speed':'mean', 
                                     'ismar1_maximum_wind_speed':'max','ismar1_significant_wave_height':'mean', 
                                     'ismar1_highest_wave_height':'max','ismar1_atmospheric_pressure':'median', 
                                     'ismar1_air_temperature':'median','ismar1_water_temperature':'median', 
                                     'ismar1_relative_humidity': 'median','ismar1_solar_radiation':'median', 
                                     'ismar1_precipitation':'sum','ismar2_precipitation':'sum', 
                                     'san_giorgio_average_wind_direction':'mean','san_giorgio_average_wind_speed':'mean', 
                                     'san_giorgio_maximum_wind_speed':'max','san_giorgio_air_temperature':'median', 
                                     'san_giorgio_relative_humidity':'median','san_giorgio_solar_radiation':'median', 
                                     'miser_average_sea_level':'mean','miser_average_wind_direction':'mean', 
                                     'miser_average_wind_speed':'mean','miser_maximum_wind_speed':'max', 
                                     'miser_significant_wave_height':'mean','miser_highest_wave_height':'max', 
                                     'salute_average_sea_level':'mean','salute_water_temperature':'median', 
                                     'faro_average_sea_level':'mean','faro_average_wind_speed':'mean', 
                                     'faro_maximum_wind_speed':'max','faro_average_wind_direction':'mean', 
                                     'astro_tide':'mean'}).reset_index()
venice
```

The column ismar2_precipitation should exactly reflect the cumulative precipitations observed at the ISMAR-CNR Platform as the documentation suggests. However, with the code below, we see that our cumulative aggregation of 'ismar1_precipitation' differs from the 'ismar2_precipitation'.\
For this reason we keep both columns.

```python
venice.resample('H',on='date').sum()['ismar1_precipitation'].equals(venice.resample('H',on='date').sum()['ismar2_precipitation'])
```
